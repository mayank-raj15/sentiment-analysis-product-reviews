# Sentimental Analysis of Amazon Product Reviews

```python
import json
import csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report, accuracy_score
```

## Initial Data Format

- Contains a review object in a single line
- Each review is discrete
- Each review has following data values
  - **reviewerID** (Unique ID of each reviewer)
  - **asin** (Amazon Standard Identification Number, which is the unique code for each product)
  - **reviwerName** (Name of the reviewer)
  - **helpful** (Contains the number of people who found review helpful, and the total people who voted)
  - **reviewText** (Text content of the review)
  - **overall** (Rating value)
  - **summary** (Title of the review)
  - **unixReviewTime** (Time of the review in UNIX format)
  - **reviewTime** (Date of the review)



## Data Cleaning

### Creating a JSON file with an object containing all reviews in an array

```python
# splitting the data in the json file into lines
with open('raw_review_data.json') as json_file:
    lines = json_file.read().splitlines()
    
# adding a comma at the end of each line
with open('review_data.json', 'w') as json_file:
    for i in range(len(lines)):
        json_file.write(lines[i])
        if i < len(lines)-1:
            json_file.write(',\n')
            
# reading the updated data from file
with open('review_data.json') as json_file:
    full_data = json_file.read()
    
# adding the reviewData wrapper & converting each review line into elements of an array
with open('review_data.json', 'w') as json_file:
    json_file.write("{\"reviewData\": [" + full_data + "]}")
```

#### JSON file


### Creating the CSV file with each review in a single row

```python
# loading the data from the final updated json file
with open('review_data.json') as json_file:
    data = json.load(json_file)
    
# getting the list of reviews
reviewData = data['reviewData']
```

```python
print(reviewData[:5])
```

```python
# writing to the csv file
with open('review_data.csv', 'w', newline='') as data_file:
    # initializing the csv writer for the current file
    csv_writer = csv.writer(data_file)
    
    # writing the heading row 
    header = ['text', 'overall', 'helpfulness']
    csv_writer.writerow(header)
    
    for review in reviewData:
        # review text
        text = review['summary'] + review['reviewText']
        
        # rating
        rating = review['overall']
        
        # calculating helpfulness
        a = review['helpful'][0]
        b = review['helpful'][1]
        helpfulness = 0 if b==0 else a/b
            
        # entering the data in csv file
        rowData = [text, rating, helpfulness]
        csv_writer.writerow(rowData)
```

### CSV file


## Sentiment Generation & Data analysis

### Disclaimer
- Sentiments are generated on the basis of overall rating.
- There might be cases when rating doesn't justify the review
- This may lead to lead to false results in boundary cases
- Especially for reviews that are neutral, there is high chance of false classification

### Sentiment Generation

```python
# reading data from csv file using Pandas
review_data = pd.read_csv('review_data.csv')

#checking if there are rows without data
review_data.isnull().sum()
```

```python
def generateSentiment(rating):
    if rating>3:
        return 'positive'
    elif rating==3:
        return 'neutral'
    else:
        return 'negative'
    
review_data['sentiment'] = review_data['overall'].apply(generateSentiment)
```

```python
review_data.head()
```

### Analysis

#### Checking the variety of sentiments

```python
review_data['sentiment'].value_counts()
```

We can see that number of positive reviews is much higher than other categories. This might effect the results later.

#### Checking whether helpfulness value is helpful or not

```python
review_data['helpfulness'].value_counts()
```

```python
pd.DataFrame(review_data.groupby('sentiment')['helpfulness'].mean())
```

```python
plt.rcParams.update({'font.size': 18})
plt.rcParams['figure.figsize'] = 16,9

# Creating dataframe and removing 0 helpfulrate records
senti_help = pd.DataFrame(review_data, columns = ['sentiment', 'helpfulness'])
senti_help = senti_help[senti_help['helpfulness'] != 0.00] 

#Plotting phase
sns.violinplot( x=senti_help["sentiment"], y=senti_help["helpfulness"])
plt.title('Sentiment vs Helpfulness')
plt.xlabel('Sentiment categories')
plt.ylabel('helpful rate')
plt.show()
```

#### Findings
- Positive reviews are slightly more helpful then neutral and negative reviews
- But negative & netural reviews have similar helpfulness value
- Also from the above plot, we can see that helpfulness value of negative & neutral reviews has same effect on the sentiment
- Hence, helpfulness value is not a good feature for classification

## Feature Extraction & Training

#### Adding Stopwords

```python
stopwords = ['a', 'am', 'about', 'an', 'and', 'are', 'as', 'at', 'be', 'been', 'but', 'by', \
             'even', 'ever', 'for', 'from', 'get', 'had', 'has', 'have', 'he', 'her', 'hers', 'his', \
             'how', 'i', 'if', 'in', 'into', 'is', 'it', 'its', 'just', 'me', 'my', 'of', 'on', 'or', \
             'see', 'seen', 'she', 'so', 'than', 'that', 'the', 'their', 'there', 'they', 'this', \
             'to', 'we', 'what', 'when', 'which', 'who', 'will', 'with', 'you']
```

### Creating Training & Testing Data

```python
X = review_data['text']
y = review_data['sentiment']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model_lr = Pipeline([('tfidf', TfidfVectorizer(stop_words=stopwords, ngram_range=(1,2))), \
                     ('lr', LogisticRegression(solver='saga'))])
```

### Training the Logistic Regression Model

```python
model_lr.fit(X_train, y_train)
```

### Testing

#### Getting predictions for Test Data

```python
predictions = model_lr.predict(X_test)
```

#### Printing Confusion Matrix

```python
cm = confusion_matrix(y_test, predictions, labels=model_lr.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model_lr.classes_)
disp.plot()
plt.show()
```

#### Printing Classification Report

```python
print(classification_report(y_test,predictions))
```

#### Accuracy Score

```python
print(accuracy_score(y_test,predictions))
```
